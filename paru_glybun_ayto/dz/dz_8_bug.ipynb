```python
print("dz_8_bug")

import random
from threading import ExceptHookArgs
import enum
from strenum import StrEnum
from typing_extensions import Self 
        
class Names(StrEnum):
    JOHN = "John Lennon"
    PAUL = "Paul McCartney"
    GEORGE = "George Harrison"
    RINGO = "Ringo Starr"


class Beetle:
    health_points: int
    name: Names

    def __init__(
        self,
        health_points: int = 100,
        name: Names = Names.JOHN,
    ) -> None:
        self.health_points = health_points
        self.name = name

    def __eq__(self, other: Self) -> bool:
        return self.health_points == other.health_points

    def __lt__(self, other: Self) -> bool:
        return self.health_points < other.health_points

    def __le__(self, other: Self) -> bool:
        return self.health_points <= other.health_points

    def __str__(self) -> str:
        return f'Beetle(name="{self.name}", hp={self.health_points!r})'

    def styling(self) -> str:
        if self.name is Names.JOHN:
            return "in Johny style"
        elif self.name is Names.PAUL:
            return "in McCartney style"
        return "without style"

    def attack(self, other: Self) -> None:
        print(f"{self.name} attacking {other.name} {self.styling()}")
        other.health_points -= 10


class BeetlesArmy:
    beetles_list: list[Beetle]
    beetles_name: Names
    beetles_max_health_points: int

    def __init__(
        self,
        beetles_name: Names,
        beetles_army_size: int = 20,
        beetles_max_health_points: int = 100,
    ):
        self.beetles_list = []
        self.beetles_name = beetles_name
        self.beetles_max_health_points = beetles_max_health_points

        for _ in range(beetles_army_size):
            beetle = Beetle(
                health_points=random.randint(1, self.beetles_max_health_points),
                name=self.beetles_name,
            )
            self.beetles_list.append(beetle)

    def __len__(self) -> int:
        return len(self.beetles_list)

    def __add__(self, other: Self) -> Self:
        if self.beetles_name != other.beetles_name:
            raise ValueError("Cannot make two different-named beetles friends")
        new_beetles_list: list[Beetle] = self.beetles_list + other.beetles_list
        new_army = self.__class__(
            beetles_army_size=1,
            beetles_name=self.beetles_name,
            beetles_max_health_points=self.beetles_max_health_points,
        )
        new_army.beetles_list = new_beetles_list
        return new_army

    def print_army_listing(self):
        for beetle in self.beetles_list:
            print(beetle)









num_arm = int(input('Создание армии жуков, введите любимое число. которое влияет на то, сколько будет армий: '))
arms = []
arm_nm = []
for i in range(num_arm):
    print(f"Какой у нас номер {i+1}")
    print("Какое имя у Армии, выберете: 1-Джош Ленон, 2-Поил Картеу, 3-Георг Хариссон, 4-Ринго Звеззда")
    chc = input('Введите цифру от 1 до 4 включительно, не больше: ')
    if chc == "1":
        nm_chc = Names.JOHN
    elif chc == "2":
        nm_chc = Names.PAUL
    elif chc == "3":
        nm_chc = Names.GEORGE
    elif chc == "4":
        nm_chc = Names.RINGO
    else:
        print("Не тот выбор, всё равно победит - John")
        nm_chc = Names.JOHN
    arm_nm.append(nm_chc)
    sz=int(input('Сколько жуков в армии: '))
    max_hp=int(input("Максимальное здоровье жуков: "))

    arm = BeetlesArmy(
        beetles_name=nm_chc,
        beetles_army_size= sz,
        beetles_max_health_points=max_hp)
    arms.append(arm)
    print(f"Армия {nm_chc} создан, Илья Лубинец, давайте их бить!!!!")

def batt_arm(armies_list):
    print("Start bitva!!!!")
    rund = 1
    while True:
        print(f" Раунд {rund} ")
        al_arm = []
        for arm in armies_list:
            arm.beetles_list = [beetle for beetle in arm.beetles_list if beetle.health_points>0]
            if len(arm.beetles_list)>0:
                al_arm.append(arm)
        if len(al_arm) <=1:
            break
        for a, att_arm in enumerate(al_arm):
            if len(att_arm.beetles_list) == 0:
                continue
                
            trg_arms = [arm for b, arm in enumerate(al_arm) if b != a and len(arm.beetles_list) > 0]
            if not trg_arms:
                continue
                
            trg_arm = random.choice(trg_arms)
            attk = random.choice(att_arm.beetles_list)
            defn = random.choice(trg_arm.beetles_list)
            print(f"\n{att_arm.beetles_name} атакует {trg_arm.beetles_name}!")
            old_hp = defn.health_points
            dm = random.randint(5, 16)
            defn.health_points -= dm
            print(f"{attk.name} бьет {defn.name} на {dm} урона!")
            print(f"Здоровье {defn.name}: {old_hp} -> {defn.health_points}")

            if defn.health_points <= 0:
                print(f" {defn.name} УМЕР, не знаю, хорошо это или нет!")
                attk.health_points += 10
                if attk.health_points > att_arm.beetles_max_health_points:
                    attk.health_points = att_arm.beetles_max_health_points
                    print(f"{attk.name} получает +10 HP за убийство! Теперь у него {attk.health_points} HP, наверное")
                trg_arm.beetles_list.remove(defn)
        
        print(f"\nСостояние после кровавийшей битвы, все возможно умерли, {rund}:")
        for arm in al_arm:
            ttl_hp = sum(beetle.health_points for beetle in arm.beetles_list)
            print(f"Армия {arm.beetles_name}: {len(arm.beetles_list)} жуков, общее HP: {ttl_hp}")
        rund += 1
    wnnr = None
    for arm in armies_list:
        if len(arm.beetles_list) > 0:
            wnnr = arm
            break
    if wnnr:
        ttl_hp=sum(beetle.health_points for beetle in wnnr.beetles_list)
        print(f" выиграл: {wnnr.beetles_name}, осталось в живых: {len(wnnr.beetles_list)}, а вот общее хп {ttl_hp}")
    else:
        print("Ничья!!!!!Все сдохли(())")





def com_arm(armies_list):
    arm_stts=[]
    for arm in armies_list:
        alv_beetles  = [b for b in arm.beetles_list if b.health_points > 0]
        ttl_hp = sum(beetle.health_points for beetle in alv_beetles)
        avg_hp = ttl_hp / len(alv_beetles) if alv_beetles else 0
        arm_stts.append({
            'name': arm.beetles_name,
            'count': len(alv_beetles),
            'total_hp': ttl_hp,
            'avg_hp': avg_hp})
        
    arm_stts.sort(key=lambda x:x['count'], reverse=True)
    for z, stts in enumerate(arm_stts):
        print(f"{z+1} {stts['name']}: {stts['count']} жуков, {stts['total_hp']} здор, среднее: {stts['avg_hp']:.1f}")



batt_arm(arms)
print("END!!((()))")
```
