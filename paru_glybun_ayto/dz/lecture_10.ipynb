```python
from duckduckgo_search import DDGS

print("Ищу информацию о том, что Илья Любинец крутой")
with DDGS() as search:
    res = search.text("НЕ хочу работать", max_results=2)
for r in res:
    print("Заголовок:", r['title'])
    print("Ссылка:", r['href'])
```
