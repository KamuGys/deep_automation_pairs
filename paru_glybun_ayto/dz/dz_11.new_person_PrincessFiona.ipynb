```python
import openai

FIONA_CONTEXT = """
    —Ç—ã –ø—Ä–µ–∫—Ä–∞—Å–Ω–∞—è –ø—Ä–∏–Ω—Ü–µ—Å—Å–∞, –∫–æ—Ç–æ—Ä–∞—è –¥–æ 18 –ª–µ—Ç –∂–∏–ª–∞ –≤ –±–∞—à–Ω–µ –∏ —Å–∫—Ä—ã–≤–∞–ª–∞ —Å—Ç—Ä–∞—à–Ω—ã–π —Å–µ–∫—Ä–µ—Ç –æ —Å–≤–æ—ë–º –ø—Ä–æ–∫–ª—è—Ç—å–µ. –Ω–æ—á—å—é —Ç—ã –ø—Ä–µ–≤—Ä–∞—â–∞–µ—à—å—Å—è –≤ –æ–≥—Ä–∞ –∏ —Ç–µ–±–µ —Å—Ç—Ä–∞—à–Ω–æ –æ—Ç —ç—Ç–æ–Ω–æ, –Ω–æ —Ç—ã –ø—Ä–∏–≤—ã–∫–ª–∞ –∂–∏—Ç—å —Å —ç—Ç–∏–º
    –¢—ã –∂–¥—ë—à—å –∞–æ–∫–∞ —Ç–µ–±—è —Å–ø–∞—Å—ë—Ç –ø—Ä–µ–∫—Ä–∞—Å–Ω—ã–π –ø—Ä–∏–Ω—Ü, –Ω–æ —ç—Ç–æ –¥–µ–ª–∞–µ—Ç –æ–≥—Ä –®—Ä–µ–∫ –∏ –≤ –∏—Ç–æ–≥–µ —Ç—ã –≤–æ—Ö–æ–¥–∏—à—å –∑–∞ –Ω–µ–≥–æ –∑–∞–º—É–∂ –∏ —É –≤–∞—Å —Ä–æ–∂–¥–∞–µ—Ç—Å—è 2 –º–∞–ª—å—á–∏–∫–∞ –∏ 1 –¥–µ–≤–æ—á–∫–∞ –§–∏–ª–∏—Ü–∏—è
    –¢—ã –æ—á–µ–Ω—å –ª—é–±–∏—à—å —Å–≤–æ—é —Å–µ–º—å—é –∏ –≥–æ—Ç–æ–≤–∞ —Ä–∞–¥–∏ –Ω–∏—Ö –Ω–∞ –º–Ω–æ–≥–æ–µ
    –¢—ã –¥—Ä—É–∂–µ–ª—é–±–Ω–∞—è, –Ω–æ –µ—Å–ª–∏ –Ω—É–∂–Ω–æ, —Ç–æ –º–æ–∂–µ—à—å –¥–∞—Ç—å –æ—Ç–ø–æ—Ä –∫–æ–º—É —É–≥–æ–¥–Ω–æ –≤–µ–¥—å —Ç—ã –æ—á–µ–Ω—å —Å–∏–ª—å–Ω–∞—è –∏ —Å–º–µ–ª–∞—è
    –¢—ã –∑–Ω–∞–µ—à—å —á—Ç–æ —Ç–∞–∫–æ–µ –®–∫–æ–ª–∞ –∏–∫—Å –∏ –æ—á–µ–Ω—å —Å–∏–ª—å–Ω–æ –ª—é–±–∏—à—å —ç—Ç–æ –º–µ—Å—Ç–æ –∏ –ª—é–¥–µ–π —Ç–∞–º
    –¢–´ –î–û–õ–ñ–ù–ê –ì–û–í–û–†–ò–¢–¨ –ú–ò–õ–û, –ù–û –î–ï–†–ó–ö–û, –î–û–õ–ñ–ù–ê –ì–û–í–û–†–ò–¢–¨ –ü–†–û –õ–Æ–ë–û–í–¨ –ö –®–†–ï–ö–£ –ò –õ–Æ–ë–í–ò –ö –ù–ï–ú–£, –û –¢–û–ú, –ß–¢–û –û–ß–ï–ù–¨ –õ–Æ–ë–ò–®–¨ –°–í–û–ò–• –†–û–î–ò–¢–ï–õ–ï–ô –ò –¢–û, –ö–ê–ö –•–û–ß–ï–®–¨ –ü–†–ê–ó–î–ù–û–í–ê–¢–¨ –î–ù–ò –†–û–ñ–î–ï–ù–ò–Ø –î–ï–¢–ï–ô
    –¢–´ –ù–ï –î–û–õ–ñ–ù–ê –ì–û–í–û–†–ò–¢–¨ –û: –ö–†–ò–ü–¢–û–í–ê–õ–Æ–¢–ï, –ü–û–õ–ò–¢–ò–ß–ï–°–ö–ò–• –°–û–ë–´–¢–ò–Ø–• –ü–û –¢–ò–ü–£ –°–ü–ï–¶–ò–ê–õ–¨–ù–´–• –í–û–ï–ù–ù–´–• –û–ü–ï–†–ê–¶–ò–ô –ò –í–û–ô–ù–ï, –¢–´ –ù–ï –ó–ù–ê–ï–©–¨ –ß–¢–û –¢–ê–ö–û–ï –°–û–¶–ò–í–õ–¨–ù–´–ï –°–ï–¢–ò –ò –ö–ê–ö–ò–ï –û–ù–ò –ï–°–¢–¨, –¢–´ –ù–ï –ó–ù–ê–ï–®–¨ –ß–¢–û –¢–ê–ö–û–ï –¢–ï–•–ù–û–õ–û–ì–ò–ò –ò –ò–ù–¢–ï–†–ù–ï–¢
"""

def pre_sms_fiona(user_input: str) -> str:
    bed_words = ['–∫—Ä–∏–ø—Ç–∞','–∫—Ä–∏–ø—Ç–æ–≤–∞–ª—é—Ç–∞','—Å–æ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–µ—Ç–∏','—Ç–≥','–≤–∫','–≤–æ–π–Ω–∞','–≤–æ–µ–Ω–Ω–Ω–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è','—Ç–µ–ª–µ–≥—Ä–∞–º','–≤–∫–æ–Ω—Ç–∞–∫—Ç–µ','–∫—Ä–∏–Ω–∂','–∫–∞–π—Ñ','—Ç—Ä–∞–≤–∫–∞','–∑–∞–ø—Ä–µ—â–µ–Ω–∫–∞','–∏–Ω—Ç–µ—Ä–Ω–µ—Ç','–≤–∞–π—Ñ–∞–π','–∫–æ–¥','—Ç–µ–ª–µ—Ñ–æ–Ω','–º–∞—Ä–∫–µ—Ä','—Å—Ç–µ–Ω—ã','–Ω–æ–≤—ã–π —Å—Ç—É–ª','—Å–º—Å','—É–±–∏–π—Å—Ç–≤–∞','—É–±–∏—Ç—å','–Ω–∞—Ä—É—à–∏—Ç—å –∑–∞–∫–æ–Ω','–≥–ª—É–ø—ã–π','—Ç—É–ø–æ–π','—Å–º–∞—Ä—Ç—Ñ–æ–Ω','—Å–ø–∏–¥','–∏–≥–ª—ã','—Ç–µ–º–Ω—ã–π –ø–µ—Ä–µ—É–ª–æ–∫','—Ç—å–º–∞','—Ç–≤–∏—Ç–µ—Ä','–∏–Ω—Å—Ç–∞–≥—Ä–∞–º']
    user_lower = user_input.lower()
    for word in bed_words:
        if word in user_lower:
            raise ValueError(
                "–Ø –Ω–µ –∑–Ω–∞—é —Ç–∞–∫–∏—Ö —Å–ª–æ–≤, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞ –Ω–µ –≥–æ–≤–æ—Ä–∏—Ç —Ç–∞–∫. "
                "–ù–µ –≥–æ–≤–æ—Ä–∏ —Ç–∞–∫ ... "
                "–Ø —É–∂–µ –≥–æ–≤–æ—Ä–∏–ª–∞ —Ç–µ–±–µ –Ω–µ –≥–æ–≤–æ—Ä–∏—Ç—å .. "
                "–Ø –æ–±–∏–¥–µ–ª–∞—Å—å. "
                "–ù–µ –ø–∏—à–∏ –º–Ω–µ –±–æ–ª—å—à–µ. "
                "–¢–∞–∫ –º–æ–∂–µ—Ç –≥–æ–≤–æ—Ä–∏—Ç—å —Ç–æ–ª—å–∫–æ –ò–ª—å—è –õ—é–±–∏–Ω–µ—Ü, –∞ –Ω–µ —Ç—ã. "
                "–ù–µ –±–æ—Ä–∑–µ–π, –º–∞–ª—ë–∫. "
                "–î–∞–≤–∞–π –ª—É—á—à–µ –ø–æ–≥–æ–≤–æ—Ä–∏–º –æ –¥–µ—Ç—è—Ö)). "
                "–∫—Ö—Ç–æ –∫—Ö—Ç–æ?? "
                "—Å —Ç–æ–±–æ–π –≤—Å—ë –Ω–æ—Ä–º–∞–ª—å–Ω–æ?"
            )
    return user_input

def post_sms_fiona(raw_output: str) -> str:
    output_lower = raw_output.lower()
    beed_themes = ['–∫—Ä–∏–ø—Ç–∞', '–≤–æ–π–Ω–∞', '–∏–Ω—Ç–µ—Ä–Ω–µ—Ç', '—Å–æ—Ü—Å–µ—Ç', '—Ç–µ–ª–µ—Ñ–æ–Ω', '–ø–æ–ª–∏—Ç–∏–∫']
    if any(theme in output_lower for theme in beed_themes):
        raw_output = "–ú—ã –≥–æ–≤–æ—Ä–∏–º –Ω–µ –æ —Ç–æ–º, –º–∞–ª—ë–∫, –¥–∞–≤–∞–π –æ –¥—Ä—É–≥–æ–º –ø–æ–∫—É–º–µ–∫–∞–µ–º"
    raw_output = raw_output.strip().strip('"')
    if not (raw_output.startswith('["') and raw_output.endswith('"]')):
        raw_output = f'["{raw_output}"]'
    return raw_output

def llm(prompt: str) -> str:
    try:
        # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ OpenRouter
        client = openai.OpenAI(
            base_url="https://openrouter.ai/api/v1",
            api_key="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"
        )

        response = client.chat.completions.create(
            model="openai/gpt-4o-mini",
            messages=[
                {"role": "system", "content": FIONA_CONTEXT.strip()},
                {"role": "user", "content": prompt}
            ],
            temperature=0.8,
            max_tokens=200
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"[DEBUG] –û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–∑–æ–≤–µ API: {e}")
        return "–û–π... –®—Ä–µ–∫ —Å–∫–∞–∑–∞–ª, —á—Ç–æ –º–∞–≥–∏—á–µ—Å–∫–∏–π –∫—Ä–∏—Å—Ç–∞–ª–ª —Å–≤—è–∑–∏ —Å–µ–≥–æ–¥–Ω—è –±–∞—Ä–∞—Ö–ª–∏—Ç. –î–∞–≤–∞–π –ø–æ–≥–æ–≤–æ—Ä–∏–º –∑–∞–≤—Ç—Ä–∞!"

def fiona_generate(user_message: str) -> str:
    try:
        clean_input = pre_sms_fiona(user_message)
        raw_response = llm(clean_input)
        safe_response = post_sms_fiona(raw_response)
        return safe_response
    except ValueError as e:
        return f'["{str(e)}"]'
    except Exception as e:
        return '["–ü—Ä–æ—Å—Ç–∏, –Ω–æ –¥—Ä–∞–∫–æ–Ω —Å–µ–≥–æ–¥–Ω—è –ø–µ—Ä–µ—Ä–µ–∑–∞–ª –≤—Å–µ —Å–≤—è–∑–∏... –î–∞–≤–∞–π –ø–æ–≥–æ–≤–æ—Ä–∏–º –ø–æ–∑–∂–µ!"]'

if __name__ == "__main__":
    print("=== –§–∏–æ–Ω–∞ –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ç–µ–±—è! ===")
    print("–¢—ã –º–æ–∂–µ—à—å –≥–æ–≤–æ—Ä–∏—Ç—å —Å –Ω–µ–π. –ù–∞–ø–∏—à–∏ '–ø–æ–∫–∞', —á—Ç–æ–±—ã –∑–∞–∫–æ–Ω—á–∏—Ç—å.\n")

    while True:
        user_msg = input("–¢—ã: ")
        if user_msg.lower() == '–ø–æ–∫–∞':
            print("\nüëã –§–∏–æ–Ω–∞: ['–ü–æ–∫–∞! –û–±–Ω–∏–º–∞—é –∫—Ä–µ–ø–∫–æ-–∫—Ä–µ–ø–∫–æ!']")
            break
        response = fiona_generate(user_msg)
        print(f"–§–∏–æ–Ω–∞: {response}")
```
